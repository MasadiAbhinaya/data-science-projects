```python
import pandas as pd

data = {
    "Product": ["Laptop", "Phone", "Tablet", "Headphones", "Smartwatch"],
    "Sales": [150, 200, 120, 180, 90],
    "Price": [70000, 30000, 20000, 5000, 15000]
}

df = pd.DataFrame(data)

df
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Product"], df["Sales"])
plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```
